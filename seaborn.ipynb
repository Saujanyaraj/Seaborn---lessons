```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

# Scatter Plot

```python
sns.scatterplot(x = "total_bill", y = 'tip', data = tips)
```

# LinePLot

```python
sns.lineplot( x = "size", y = "tip" , data = tips)
```

# Dist PLot

```python
sns.distplot(tips['tip'])
```

```python
## Bar PLot
```

```python
sns.barplot( x = "day", y ="tip", data = tips)
```

```python
## Kernel Density PLot
```

```python
sns.kdeplot(tips['tip'])
```

```python
## Reg PLot
```

```python
sns.regplot( x = 'total_bill', y = 'tip', data = tips)
```

```python
## Boxplot
```

```python
sns.boxplot( x = 'size', y = 'tip', data = tips)
```

```python
## Cat plot( Category plot)
```

```python
sns.catplot( x = "day", y = "tip", data = tips)
```

```python
## Joint PLot (Draw plot of two variables with bivariate and univariate graph )
```

```python
sns.jointplot( x = 'total_bill', y = 'tip', data = tips, kind = 'reg')
```

```python
## Count Plot ( Feature against no. of observations/ ocurenece)
```

```python
sns.countplot ( x = 'smoker', data = tips)
```

```python
## Pair Plot ( Visualize relationship between all fetures(numerical) in data seta at once)
```

```python
sns.pairplot(tips)
```

```python
## Relation Plot ( relationship between two variables with sematic mapping of subsets)
```

```python
sns.relplot( x = 'total_bill', y = 'tip', col = 'sex', hue = 'smoker', style = 'size', data = tips)
```

```python
## Heat Map
```

```python
sns.heatmap(tips.corr(), linecolor = 'white', linewidths = 2, annot = True)
```
